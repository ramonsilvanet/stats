```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data/images_per_category.csv', header = 1 , quoting=2, names=["quant. de imagens"], dtype=np.int64)
data.plot(kind="hist", figsize=(16,9))
plt.ylabel('Categorias')
plt.xlabel('Quantidade de imagens')
plt.title("Distribuição de imagens por categoria")
plt.show()
```

A categoria que mais possui imagens associadas, possui **45174** imagens

A categoria que menos possui imagens associadas, possui **128** imagens

```python
data2 = pd.read_csv('data/categories_per_image.csv', header = 1 , quoting=2, names=["quat. de categorias"], dtype=np.int64)
data2.plot(kind="hist", figsize=(16,9))
plt.ylabel('Quantidadde de imagens')
plt.xlabel('Número de Categorias')
plt.title("Quantidade de Categorias por imagem")
plt.show()
```

O quantidade mínima de categorias que uma imagem possui é **1** e a quantidade máxima é **18**
